```python
# iris_eda.ipynb

import pandas as pd
import matplotlib.pyplot as plt

# Load data
url = "https://bit.ly/4nejNue"
df = pd.read_csv(url)

# Show first 5 rows
print("üìå First 5 Rows:")
print(df.head())

# Basic Statistics
print("\nüìä Describe:")
print(df.describe())

print("\nüìà Mean:\n", df.mean(numeric_only=True))
print("\nüìè Median:\n", df.median(numeric_only=True))
print("\nüìâ Std:\n", df.std(numeric_only=True))

# Histograms for each feature
df.hist(figsize=(10, 8), edgecolor='black')
plt.suptitle("Histogram of All Features", fontsize=16)
plt.tight_layout()
plt.show()
```
